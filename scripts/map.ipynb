```python
import skimage.transform
import skimage.measure
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 100
prob = np.random.rand(size, size)
img = np.zeros([size, size])
img[prob>0.95] = 100
print(img)
plt.imshow(img, cmap='gray')

unique, counts = np.unique(img, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
rescaled = skimage.transform.resize(img, (size/2, size/2), anti_aliasing=False)
plt.imshow(rescaled, cmap='gray')

unique, counts = np.unique(rescaled, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
pooling = skimage.measure.block_reduce(img, (2,2), func=np.max)
plt.imshow(pooling, cmap='gray')
```
